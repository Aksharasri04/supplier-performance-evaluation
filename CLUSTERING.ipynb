```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import numpy as np
```

```python
data = pd.read_csv('pca.csv')
data
```

```python
from mpl_toolkits.mplot3d import Axes3D

k_2 = KMeans(n_clusters=2, random_state=42)
k_2.fit(data)  
labels_2 = k_2.labels_
score_2 = silhouette_score(data, labels_2)  
centroids_2 = k_2.cluster_centers_

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(data.iloc[:, 0], data.iloc[:, 1], data.iloc[:, 2], c=labels_2, cmap='viridis', marker='o', s=50)
ax.scatter(centroids_2[:, 0], centroids_2[:, 1], centroids_2[:, 2], c='red', marker='X', s=200, label='Centroids')


ax.set_title('K-Means Clustering with k=2 (3D)')
ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.set_zlabel('Feature 3')
ax.legend()
plt.show()

plt.figure(figsize=(8, 6))
plt.scatter(data.iloc[:, 0], data.iloc[:, 1], c=labels_2, cmap='viridis', marker='o', s=50)
plt.scatter(centroids_2[:, 0], centroids_2[:, 1], c='red', marker='X', s=200, label='Centroids')
plt.title('K-Means Clustering with k=2')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()

print(f"Silhouette Score for k=2: {score_2}")
```

```python
silhouette_scores = []

for k in range(2, 7):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data) 
    labels = kmeans.labels_
    score = silhouette_score(data, labels)  
    silhouette_scores.append(score)

plt.figure(figsize=(8, 6))
plt.plot(range(2, 7), silhouette_scores, marker='o', linestyle='-', color='b')
plt.title('Silhouette Score for K-Means Clustering (k=2 to k=6)')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.xticks(range(2, 7))
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score

Z = linkage(data, method='ward')

plt.figure(figsize=(14, 10), dpi=120)
dendrogram(
    Z,
    labels=data.index.astype(str),  
    orientation='top',              
    leaf_font_size=10,            
    truncate_mode='lastp',          
    p=30                           
)
plt.title('Hierarchical Clustering Dendrogram', fontsize=14)
plt.xlabel('Distance', fontsize=12)
plt.ylabel('Data Point Index', fontsize=12)
plt.tight_layout()
plt.show()

max_d = 7.0  
clusters = fcluster(Z, max_d, criterion='distance')
if len(set(clusters)) > 1:
    score = silhouette_score(data, clusters)
    print("Silhouette Score for Hierarchical Clustering:", score)
else:
    print("Not enough clusters to compute the silhouette score.")
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps=0.5, min_samples=5) 
clusters = dbscan.fit_predict(data)
data['cluster'] = clusters
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

unique_clusters = np.unique(clusters)
for cluster in unique_clusters:
    mask = clusters == cluster
    label = f"Cluster {cluster}" if cluster != -1 else "Noise"
    ax.scatter(data[mask].iloc[:, 0], data[mask].iloc[:, 1], data[mask].iloc[:, 2], label=label)

ax.set_xlabel('Component 1')
ax.set_ylabel('Component 2')
ax.set_zlabel('Component 3')
ax.set_title('DBSCAN Clustering in 3D')

plt.show()
cluster_labels = clusters[clusters != -1]  
if len(np.unique(cluster_labels)) > 1:
    score = silhouette_score(data[clusters != -1], cluster_labels)
    print("Silhouette Score (excluding noise):", score)
else:
    print("Not enough clusters (excluding noise) to compute a silhouette score.")
```
